```python
#Aquí verificamos si ganó el equipo local
def checkhomewon(x):
    if(x =='H'):
        return 1
    else:
        return 0
    
```

```python
data['HomeWon']=data['FTR'].apply(checkhomewon)
```

```python
#Separamos la variable objetivo
X_all = data.drop(['FTR'],1)
y_all = data['FTR']

#Estandarizamos los datos
from sklearn.preprocessing import scale


cols = [['HTGD','ATGD','HTP','ATP','DiffLP']]
for col in cols:
    X_all[col] = scale(X_all[col])
    
def preprocess_features(X):
    ''' Procesamos los datos y convertimos las variables categoricas en dummys. '''
    
    # Iniciamos un nuevo DataFrame
    output = pd.DataFrame(index = X.index)

    # Sacamos información de cada columna 
    for col, col_data in X.iteritems():

        # Si la información es categorica, la convertimos en dummy
        if col_data.dtype == object:
            col_data = pd.get_dummies(col_data, prefix = col)
                    
        # Obtenemos los datos
        output = output.join(col_data)
    
    return output

X_all = preprocess_features(X_all)
```
